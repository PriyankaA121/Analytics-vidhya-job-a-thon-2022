```python
!rm Students.db
def generate_DB() -> None:

    # Import required libraries
    import sqlite3
    
    # Connect to SQLite database
    # New file created if it doesn't already exist
    conn = sqlite3.connect('Students.db')
    
    # Create cursor object
    cursor = conn.cursor()
    
    # Create and populate tables
    cursor.executescript('''
    CREATE TABLE Advisor(
    AdvisorID INTEGER NOT NULL,
    AdvisorName TEXT NOT NULL,
    AdvisorPost TEXT NOT NULL,
    PRIMARY KEY(AdvisorID)
    );
    
    CREATE TABLE Student(
    StudentID NUMERIC NOT NULL,
    StudentName TEXT NOT NULL,
    StudentMarks INTEGER,
    AdvisorID INTEGER,
    FOREIGN KEY(AdvisorID) REFERENCES Advisor(AdvisorID),
    PRIMARY KEY(StudentID)
    );

    CREATE TABLE City(
    CityName TEXT NOT NULL,
    StudentID NUMERIC NOT NULL,
    FOREIGN KEY(StudentID) REFERENCES Student(StudentID)
    );
    
    INSERT INTO Advisor(AdvisorID, AdvisorName, AdvisorPost) VALUES
    (1,"John Paul", "Junior"), 
    (2,"Anthony Roy", "Junior"), 
    (3,"Raj Shetty", "Junior"),
    (4,"Sam Reeds", "Senior"),
    (5,"Arthur Clintwood", "Senior");
    
    INSERT INTO Student(StudentID, StudentName, StudentMarks, AdvisorID) VALUES
    (501,"Student1",88,1),
    (502,"Student2",82,1),
    (503,"Student3",95,3),
    (504,"Student4",98,2),
    (505,"Student5",100,4),
    (506,"Student6",75,2),
    (507,"Student7",60,2),
    (508,"Student8",90,3),
    (509,"Student9",50,NULL),
    (510,"Student10",67,1);

    INSERT INTO City(CityName, StudentID) VALUES
    ("Chennai", 501),
    ("Mumbai", 503),
    ("Mumbai", 504),
    ("Pune", 506),
    ("Delhi", 508), 
    ("Delhi", 509);
    
    ''')
    
    #Commit changes to database
    conn.commit()
    
    # Closing the connection
    conn.close()


def sql_query_to_pd(sql_query_string: str, db_name: str ='Students.db') -> pd.DataFrame:
    '''Execute an SQL query and return the results as a pandas dataframe

    Args:
        sql_query_string (str): SQL query string to execute
        db_name (str, optional): Name of the SQLITE Database to execute the query in.
                                 Defaults to 'default.db'.

    Returns:
        pd.DataFrame: Results of the SQL query in a pandas dataframe
    '''    
    # Step 1: Connect to the SQL DB
    con = sqlite3.connect(db_name)

    # Step 2: Execute the SQL query
    cursor = con.execute(sql_query_string)

    # Step 3: Fetch the data and column names
    result_data = cursor.fetchall()
    cols = [description[0] for description in cursor.description]

    # Step 4: Close the connection
    con.close()

    # Step 5: Return as a dataframe
    return pd.DataFrame(result_data, columns=cols)

generate_DB()
```
